```python
import urllib.request
import sys
import pandas as pd
```

```python
blah = pd.read_csv("bird_data_indonesia.csv", encoding = "ISO-8859-1")
```

```python
blah.describe()
```

```python
sounds = blah[["id", "file"]]
```

```python
ids = set()
with open("xeno-canto-borneo.txt") as f:
    l = f.readlines()
    for item in l:
        uniqueID = item.split(":")[0].lstrip()
        uniqueID = uniqueID.split("C")[1]
        ids.add(uniqueID)
```

```python
print(ids)
```

```python
for index, row in blah.iterrows():
    recordingID = row['id']
    if recordingID.isdigit() and recordingID in ids:
        download = "http:" + row['file']
        path = "/Volumes/Seagate Por/XenoCantoSamples/"
        try:
            urllib.request.urlretrieve(download, path + recordingID + ".mp3")
        except Exception as e:
            print("Exception with: "+recordingID)
            print(e)

```
